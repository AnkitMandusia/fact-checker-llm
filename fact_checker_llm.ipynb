```python
%%writefile app.py
import spacy
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import faiss
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import json
import streamlit as st
import torch
import os
import re

# 1. Claim/Entity Detection
nlp = spacy.load("en_core_web_sm")

def extract_claims(text):
    """
    Extract key claims or entities from input text using spaCy.
    Returns a list of extracted claims or entities.
    """
    doc = nlp(text)
    claims = []
    for sent in doc.sents:
        claim = " ".join([token.text for token in sent if token.pos_ in ["NOUN", "VERB", "ADJ", "PROPN"]])
        if claim.strip():
            claims.append(claim.strip())
    return claims if claims else [text]

# 2. Trusted Fact Base
def create_fact_base(csv_path="facts.csv"):
    """
    Load a fact base from a CSV file or create a default one if not found.
    Returns a list of statements.
    """
    if os.path.exists(csv_path):
        df = pd.read_csv(csv_path)
        facts = df["fact"].tolist()
    else:
        facts = [
            "The Indian government announced free electricity for farmers in select states starting July 2025.",
            "The Ministry of Power introduced subsidies for solar panels for farmers in 2024.",
            "No nationwide free electricity policy for farmers was announced as of June 2025.",
            "The Indian government launched a rural electrification scheme in 2023.",
            "The Ministry of Agriculture announced irrigation support for farmers in 2025.",
            "Solar energy subsidies were extended to rural areas in 2024.",
            "The government plans to electrify all rural households by 2026.",
            "Free electricity for farmers was piloted in Gujarat in 2024.",
            "No policy for free electricity nationwide was confirmed as of May 2025.",
            "The Indian government increased funding for renewable energy in 2025.",
        ] + [f"Sample fact {i} about Indian government policies." for i in range(20)]
        pd.DataFrame({"fact": facts}).to_csv(csv_path, index=False)
    return facts

# 3. Embedding & Retrieval System
def create_vector_db(facts):
    """
    Create a FAISS vector database from a list of facts.
    Returns the FAISS index, embedding model, and fact list.
    """
    model = SentenceTransformer('all-MiniLM-L6-v2')
    embeddings = model.encode(facts, show_progress_bar=True)
    dimension = embeddings.shape[1]
    index = faiss.IndexFlatL2(dimension)
    index.add(embeddings)
    return index, model, facts

def retrieve_facts(claim, index, model, facts, k=3):
    """
    Retrieve top-k similar facts for a given claim.
    Returns a list of retrieved facts, their distances, and max distance for vagueness check.
    """
    claim_embedding = model.encode([claim])
    distances, indices = index.search(claim_embedding, k)
    retrieved_facts = [facts[idx] for idx in indices[0]]
    return retrieved_facts, distances[0],float(max(distances[0]))

# 4. LLM-Powered Comparison
def generate_verdict(claim, retrieved_facts, max_distance, vagueness_threshold=1.5):
    """
    Generate a verdict by comparing the claim with retrieved facts.
    Includes vagueness filtering based on max_distance.
    """
    # If too vague
    if max_distance > vagueness_threshold:
        return {
            "verdict": "Unverifiable",
            "reasoning": "The claim is too vague or unrelated to the fact base.",
            "confidence": float(0.3)
        }



    pipe = pipeline("text-generation", model="google/flan-t5-large")

    # Stronger prompt to enforce JSON structure
    prompt = f"""
You are a fact-checking assistant. Compare the claim with the retrieved facts and respond only in valid JSON with these keys:
- "verdict": one of ["True", "False", "Unverifiable"]
- "reasoning": your explanation
- "confidence": float between 0 and 1

Claim: {claim}

Facts:
{json.dumps(retrieved_facts, indent=2)}

Output example:
{{
  "verdict": "True",
  "reasoning": "The claim matches multiple facts.",
  "confidence": 0.91
}}
"""

    try:
        # Generate response
        response = llm(prompt, max_new_tokens=300, do_sample=False)[0]['generated_text']

        # Debug print (optional)
        # print("LLM raw response:", response)

        # Try parsing JSON block from response
        json_block = re.search(r'\{.*?\}', response, re.DOTALL)
        if json_block:
            try:
                # Primary attempt
                verdict = json.loads(json_block.group())
            except json.JSONDecodeError:
                # Fallback using ast
                import ast
                verdict = ast.literal_eval(json_block.group())

            # Adjust confidence with retrieved distance
            verdict["confidence"] = float(min(verdict.get("confidence", 0.5), 1.0 - max_distance / 2.0))
            return verdict
    except Exception as e:
        print("Parsing error:", e)

    # Final fallback
    return {
        "verdict": "Unverifiable",
        "reasoning": "Could not parse LLM response or no clear match found.",
        "confidence": round(0.5 - max_distance / 3.0, 3)
    }


# 5. Main Fact-Checking Pipeline
def fact_check(text):
    """
    Run the full fact-checking pipeline.
    Returns a JSON object with verdict, evidence, and reasoning.
    """
    claims = extract_claims(text)
    main_claim = claims[0]

    facts = create_fact_base()
    index, model, fact_list = create_vector_db(facts)

    retrieved_facts, distances, max_distance = retrieve_facts(main_claim, index, model, fact_list)

    verdict = generate_verdict(main_claim, retrieved_facts, max_distance)

    output = {
        "verdict": verdict["verdict"],
        "reasoning": verdict["reasoning"],
        "confidence": verdict["confidence"],
        "evidence": retrieved_facts,
    }
    return output

# 6. Streamlit UI
def run_streamlit():
    """
    Run a Streamlit UI for the fact-checking system.
    """
    st.title("LLM-Powered Fact Checker")
    st.write("Enter a claim or statement to verify its accuracy.")

    input_text = st.text_area("Input Claim", "The Indian government has announced free electricity to all farmers starting July 2025.")

    if st.button("Verify"):
        with st.spinner("Checking facts..."):
            result = fact_check(input_text)
            st.json(result)
            st.write("**Was this helpful?**")
            if st.button("Yes"):
                st.write("Thank you for your feedback!")
            if st.button("No"):
                st.write("Sorry, we'll improve!")

if __name__ == "__main__":
    sample_input = "The Indian government has announced free electricity to all farmers starting July 2025."
    result = fact_check(sample_input)
    print(json.dumps(result, indent=2))

    # Save sample input/output
    with open("sample_input.txt", "w") as f:
        f.write(sample_input)
    with open("sample_output.json", "w") as f:
        json.dump(result, f, indent=2)

    # Run Streamlit
    run_streamlit()
```
